OBJECTIVE: Create the Decision Tree classifier and visualize it graphically.

The purpose is if we feed any new data to this classifier, it would be able to
predict the right class accordingly.



```python
# Importing required python libraries
import sklearn.datasets as datasets
import pandas as pd
from sklearn.tree import DecisionTreeClassifier 
from sklearn import tree
from sklearn import datasets
```

```python
# Loading the iris dataset
iris=datasets.load_iris()

# Forming the iris dataframe
df=pd.DataFrame(iris.data, columns=iris.feature_names)

y=iris.target
```

# Defining the Decision Tree Algorithm

```python
# Defining the decision tree algorithm
dtree=DecisionTreeClassifier(random_state=1234)
dtree.fit(df,y)
```

```python
# Displaying textual representation of tree
text_representation = tree.export_text(dtree)
print(text_representation)
```

# Visualizing the Decision Tree for better understanding

```python
# Import necessary libraries for graph viz
from six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus

# Visualize the graph
dot_data = StringIO()
export_graphviz(dtree, out_file=dot_data, feature_names=iris.feature_names,filled=True,rounded=True,special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```
